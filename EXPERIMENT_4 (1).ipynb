'''Problem Statement 1: Unique Event participants

A college event has a list of stufents who registered multiple time by mistake.

Wite a python program using set to to store participant Names and display only unique participants. '''

```python
registered_participants = ["ANSH", "DIVIJ", "TRUMP", "ANSH", "CLINTON", "DIVIJ", "EPSTIEN"]

unique_participants = set(registered_participants)

print("Original Registered Participants (with duplicates):")
print(registered_participants)
print("\nUnique Participants:")
print(unique_participants)
```

'''Problem Statement 2: Three students choose elective subjects.

Find the common subjects chosen by all students using set operations.

concept used: Set intersection (&)'''

```python
# Elective subjects chosen by three stud
stud1_subjects = {"Physics", "Chemistry", "Math", "Computer Science"}
stud2_subjects = {"Math", "Biology", "Chemistry", "English"}
stud3_subjects = {"Computer Science", "Math", "Physics", "History"}

# Find common subjects using set intersection
common_subjects = stud1_subjects & stud2_subjects & stud3_subjects

print("Student 1's subjects:", stud1_subjects)
print("Student 2's subjects:", stud2_subjects)
print("Student 3's subjects:", stud3_subjects)
print("\nCommon subjects chosen by all students:", common_subjects)
```

'''Problem Statement 3: One set contains student in the cricket club another in the football club.

Write a python program to find students who are:

1- In both club

2-In only one club

```python
# Students in the Cricket Club
cricket_club = {"ANSH", "Heisenberg", "GUS", "JESSE", "SAUL"}

# Students in the Football Club
football_club = {"GUS", "MIKE", "Heisenberg", "WALTER", "HANK"}

print("Cricket Club Members:", cricket_club)
print("Football Club Members:", football_club)

# 1- Students in both clubs (Intersection)
students_in_both_clubs = cricket_club.intersection(football_club)
print("\nStudents in both clubs:", students_in_both_clubs)

# 2- Students in only one club (Symmetric Difference)
students_in_only_one_club = cricket_club.symmetric_difference(football_club)
print("Students in only one club:", students_in_only_one_club)
```

'''Problem Statement 4: Absent students

Given a set of all students and a set of present students, Write a python program to find all the absent students.

```python
all_students = {"ANSH", "DIVIJ", "NAMAN", "NEIL", "MODI"}
present_students = {"MODI", "ANSH", "DIVIJ"}

absent_students = all_students - present_students

print("All Students:", all_students)
print("Present Students:", present_students)
print("Absent Students:", absent_students)
```
